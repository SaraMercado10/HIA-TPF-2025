```python
import pandas as pd
from sqlalchemy import create_engine

# Credenciales de tu BD (usando los valores de tu docker-compose.yml)
DB_USER = 'root'
DB_PASS = 'root'
DB_HOST = 'mysql-maestro' # <--- ¡IMPORTANTE! Nombre del servicio en Docker Compose
DB_NAME = 'tpf_db_d'

# Crea la cadena de conexión
SQLALCHEMY_DATABASE_URL = f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}/{DB_NAME}"
engine = create_engine(SQLALCHEMY_DATABASE_URL)

print("Conexión a MySQL establecida.")

usuarios = 'usuarios' 

# Extraer los datos para el análisis
try:
    # Consulta que extrae los datos de tu tabla real
    query = f"SELECT * FROM {usuarios};" 
    df = pd.read_sql(query, con=engine)

    # Generar el resumen estadístico de los datos (tu informe de desempeño)
    reporte_promedio = df.describe()

    print(f"--- Informe Estadístico de la Tabla '{usuarios}' ---")
    print(reporte_promedio)

    # Guardar el informe como archivo CSV
    df.to_csv(f'/home/jovyan/work/informe_{usuarios}.csv')
    print(f"\nInforme guardado en informe_{usuarios}.csv")

except Exception as e:
    print(f"Error al ejecutar la consulta en la tabla real: {e}")
```

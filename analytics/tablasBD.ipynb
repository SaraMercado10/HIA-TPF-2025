```python
import pandas as pd
from sqlalchemy import create_engine

# Credenciales de tu BD (usando los valores de tu docker-compose.yml)
DB_USER = 'root'
DB_PASS = 'root'
DB_HOST = 'mysql-maestro' # <--- ¡IMPORTANTE! Nombre del servicio en Docker Compose
DB_NAME = 'tpf_db_d'

# Crea la cadena de conexión
SQLALCHEMY_DATABASE_URL = f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}/{DB_NAME}"
engine = create_engine(SQLALCHEMY_DATABASE_URL)

print("Conexión a MySQL establecida.")

try:
    # Consulta estándar de MySQL para obtener todas las tablas en el esquema actual
    tables_query = "SHOW TABLES;"
    
    # Ejecuta la consulta y lee el resultado en un DataFrame de Pandas
    existing_tables = pd.read_sql(tables_query, con=engine)
    
    print("--- Tablas encontradas en la BD 'tpf_db_d' ---")
    print("Elige una de estas tablas para tu análisis:")
    print(existing_tables)
    
    # Opcional: Mostrar la primera tabla encontrada para un ejemplo rápido
    if not existing_tables.empty:
        first_table = existing_tables.iloc[0, 0]
        print(f"\n--- Ejemplo: Primeros 5 registros de la tabla '{first_table}' ---")
        
        preview_query = f"SELECT * FROM {first_table} LIMIT 5;"
        preview_data = pd.read_sql(preview_query, con=engine)
        print(preview_data)

except Exception as e:
    print(f"Error al conectar o listar tablas: {e}")
```
